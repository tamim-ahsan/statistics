# 1. Spectral decomposition

```r
mat <- matrix(data = c(3, 6,
                       2, 5),
              nrow = 2)
E <- eigen(x = mat)
E
```

```r
evalues <- E$values # eigen value
evectors <- E$vectors # eigen vector
```

```r
# Verify multiplication for the first eigvalue-eigenvector pair
# matrix x eigenvector = eigenvalue x eigenvector
cat("Matrix and eigenvector\n---\n")
cat(mat %*% evectors[,1], "\n---\n")
cat("eigenvalue and eigenvector\n---\n")
cat(evalues[1] * evectors[,1])
```

```r
# Verify multiplication for the second eigvalue-eigenvector pair
# matrix x eigenvector = eigenvalue x eigenvector
cat("Matrix and eigenvector\n---\n")
cat(mat %*% evectors[,2], "\n---\n")
cat("eigenvalue and eigenvector\n---\n")
cat(evalues[2] * evectors[,2])
```

```r
# Trace
sum(diag(mat)) == sum(evalues)
```

```r
mat_svd <- svd(x = mat)
mat_svd
```
